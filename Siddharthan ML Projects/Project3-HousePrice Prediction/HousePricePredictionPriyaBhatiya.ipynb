### Prepare data

```python
dataset = pd.DataFrame(data=housing.data, columns=housing.feature_names)
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset['target'] = housing.target
dataset.head()
```

```python
dataset.rename(columns={'target': 'Price'}, inplace=True)
dataset.head()
```

```python
dataset.info()
```

### Exploratory Data Analysis

```python
dataset.corr()
```

```python
dataset.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(dataset.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
```

```python
sns.pairplot(dataset, diag_kind='kde', markers='o')
```

```python
# Box plot to find outliers
fig, ax = plt.subplots(figsize=(15,15))
sns.boxplot(data=dataset, ax=ax)
plt.title('Box Plot of Features')
```

```python
# split the data into independent and dependent features
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
# split the dataset into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```
