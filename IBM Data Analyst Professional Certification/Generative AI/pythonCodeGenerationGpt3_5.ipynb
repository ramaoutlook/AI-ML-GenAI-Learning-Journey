```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np
from scipy.stats import pearsonr
```

```python
df.dtypes
```

```python
# Convert Screen Size from cm to inches
df["Screen_Size_inch"] = df["Screen_Size_cm"] * 0.393701

# Convert Weight from kg to pounds
df["Weight_pounds"] = df["Weight_kg"] * 2.20462

# Display updated DataFrame
df.head()
```

```python
#df = df.dropna(subset=["CPU_frequency", "Screen_Size_inch", "Weight_pounds", "Price"])
```

```python
# Set the style for plots
sns.set_style("whitegrid")

# Regression Plots
regression_features = ["CPU_frequency", "Screen_Size_inch", "Weight_pounds"]
plt.figure(figsize=(15, 5))
for i, feature in enumerate(regression_features, 1):
    plt.subplot(1, 3, i)
    sns.regplot(x=df[feature], y=df["Price"])
    plt.title(f"{feature} vs Price")

plt.tight_layout()
plt.show()
```

```python
# Box Plots
box_features = ["Category", "GPU", "OS", "CPU_core", "RAM_GB", "Storage_GB_SSD"]
plt.figure(figsize=(15, 10))
for i, feature in enumerate(box_features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[feature], y=df["Price"])
    plt.xticks(rotation=45)
    plt.title(f"{feature} vs Price")

plt.tight_layout()
plt.show()
```

```python
# Assuming df is your DataFrame with numerical columns including "Price"
numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()
numerical_features.remove("Price")  # Remove "Price" itself from comparison

# Initialize an empty list to store results
correlation_results = []

# Calculate Pearson correlation coefficient and p-value for each numerical feature
for feature in numerical_features:
    pearson_corr, p_value = pearsonr(df[feature], df["Price"])
    correlation_results.append([feature, pearson_corr, p_value])

# Create a DataFrame to display results
correlation_df = pd.DataFrame(correlation_results, columns=["Feature", "Pearson_Correlation", "P_Value"])

# Display the final DataFrame
print(correlation_df)
```

```python
# Group the attributes and create a pivot table
pivot_df = df.pivot_table(values="Price", index="GPU", columns="CPU_core", aggfunc=np.mean)

# Plot pcolor plot
plt.figure(figsize=(10, 6))
plt.pcolor(pivot_df, cmap="coolwarm", edgecolors='k')

# Add color bar
plt.colorbar(label="Mean Price")

# Add labels
plt.xticks(np.arange(0.5, len(pivot_df.columns), 1), pivot_df.columns, rotation=45)
plt.yticks(np.arange(0.5, len(pivot_df.index), 1), pivot_df.index)

plt.xlabel("CPU Core")
plt.ylabel("GPU")
plt.title("Pcolor Plot of Price Pivot Table")

plt.show()
```
