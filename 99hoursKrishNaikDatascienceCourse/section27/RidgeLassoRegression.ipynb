## Data Cleaning

```python
dataset.isnull().sum()
```

```python
dataset.isnull().any(axis=1)
```

```python
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on REgion from 122th index, we can make a new column based on the Region

1: Bejaia Rgion Dataset
2: Sidi-Bel Abbes Region Dataset

Add new column with region

```python
dataset.loc[:122, "Region"]=0
dataset.loc[122:, "Region"]=1
df = dataset
```

```python
df.info()
```

```python
df[['Region']]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
 ## fix spaces in column names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
## Change the required columns as integer datatype
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']]=df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype('int')
```

```python
df.info()
```

```python
df.columns
```

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!= 'Classes':
        df[i] = df[i].astype('float')
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

```python
## Monthly fire Analysis
dftemp = df.loc[df['Region']==1]
plt.subplots(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.ylabel('Number of Fires', weight = 'bold')
plt.title("Fire Analysiss for Siri - Bel REgion")
plt.show()
```

```python
## Monthly fire Analysis
dftemp = df.loc[df['Region']==0]
plt.subplots(figsize=(13, 6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.ylabel('Number of Fires', weight = 'bold')
plt.title("Fire Analysiss for Brejai REgion")
plt.show()
```

It is observed the August and September had the most number of forest fires for both regions. and from the above plot of months, we can understand few thins
