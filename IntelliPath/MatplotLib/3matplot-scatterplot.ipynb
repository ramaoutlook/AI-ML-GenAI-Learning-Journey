```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating the dataser
a =[10,20,30,40,50,60,70,80]
b =[5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

#creating the scatter plot
plt.scatter(a,b)
plt.scatter(a,x)
plt.show()
```

```python
#customizing the scatter plot

plt.scatter(a,b, c='g', s=200, facecolor='y', marker='o', alpha=0.5)
plt.scatter(a,x, c='r', s=400, facecolor='x', marker='4', alpha=1)

plt.legend(['b', 'x'], loc='best')
plt.title("Scatter Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.savefig('ScatterPlot.png')
plt.grid(True)
plt.show()
```
