```python
# line Chart
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# defining the dataset
x=np.arange(0,10, 0.1)
y = 2*x + 5
 #ploting the dataploints
plt.plot(x, y)
plt.show()
```

## Customizing the Line Plot        

```python
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
x=np.arange(0,10, 1)
y = 2*x + 5
fig = plt.figure(figsize=(10,5))
plt.plot(x,y,linewidth=2.0, linestyle=":", color="r", alpha = 0.5, marker = "o")
plt.legend(['line1'], loc='best')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x= np.arange(0, 10, 1)
y1 = 2*x +5
y2=3*x + 10

#create two or more plots in a figure
plt.subplot(1,2,1)
#plotting the points
plt.plot(x, y1)
plt.title("Graph1")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Graph2")
plt.show()
```
