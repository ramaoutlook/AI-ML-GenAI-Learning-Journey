# Pandas Groupby function

### Pandas groupby is a function for grouping data objects into Series (Columns) or DataFrames (a group of Series) based on particular indicators. In simpler terms, group by in Python makes the management of datasets easier since you can put related records into groups

```python
import numpy as np
import pandas as pd

dfg = pd.DataFrame({
    "Employee":["Susan", "elina", "jordan", "Charles", "David", "Charles", "Julia", "Bart"],
    "City" : ["London", "London", "Philadelphia", "London", "London", "Philadelphia", "London","Philadelphia"],
    "Product" :[20,40,18,24,37,40,44,20],
    "Hours": [24,40,50,36,54,44,41,35]
})
dfg
```

```python
dfg.groupby(['Employee']).sum()
```

```python
# count how many employee are from same city
dfg.groupby(['City']).count()
```

```python
dfg.groupby('Employee')['Hours'].sum()
```

```python
dfg.groupby('Employee')['Hours'].sum().reset_index()
```

```python
#sort based on hours value
dfg.groupby('Employee')['Hours'].sum().reset_index().sort_values(by="Hours")
```

```python
# Excluding non-numeric columns
numeric_cols = dfg.select_dtypes(include=[np.number])

# Adding the 'Employee' column back to the numeric columns DataFrame
numeric_cols = pd.concat([dfg['Employee'], numeric_cols], axis=1)

# Grouping by 'Employee' and calculating the mean
result = numeric_cols.groupby('Employee').mean().reset_index()
print(result)
```
