```python
lilst_of_tuples = [('sf', 2012), ('phx', np.nan), ('phx', 2005), ('chi', 2009)]
df = pd.DataFrame(lilst_of_tuples, columns=['city', 'year'])
df['country']='Bharat'
df['continent'] = 'Asia'
df
```

```python
#you can remove a column from a dataframe with the .drop() method by referencing its label and axis
#axis 0 = rows
# axis 1 = columns
df2 = df.drop('country', axis=1, inplace = False)
df2
```

```python
df3 = pd.DataFrame(df2)
print(df3)
del(df3['continent'])
df3
```
