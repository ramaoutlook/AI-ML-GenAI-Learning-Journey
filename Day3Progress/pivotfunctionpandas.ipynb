```python
pivot_table = pd.pivot_table(df, index=['Name'], values=['Quantity', 'Price'], aggfunc='mean')
print(pivot_table)
```

```python
pd.pivot_table(df, index=['Rep', 'Manager'], values=['Price'], aggfunc='sum')
```

```python
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'])
```

```python
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'], aggfunc=['mean'])
```

```python
pd.pivot_table(df, index=['Manager', 'Status'], columns=['Product'], values=['Quantity', 'Price'],
               aggfunc={'Quantity':len, 'Price':['sum', 'mean']},fill_value=0)
```

```python
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'], columns=['Product'],aggfunc=['sum'])
```

```python
pd.pivot_table(df, index=['Manager', 'Rep', 'Product'], values=['Price', 'Quantity'], aggfunc=['sum'], fill_value=0, margins=True)
```

```python
pd.pivot_table(df, index=['Manager', 'Status'],columns=['Product'],values=['Quantity', 'Price'],
               aggfunc={'Quantity':len, 'Price':['sum', 'mean']}, fill_value=0)
```
